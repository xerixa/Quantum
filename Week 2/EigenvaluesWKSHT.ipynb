```julia
using LinearAlgebra

A = randn(2,2) + im * randn(2,2)

H = A + A'

FA = eigen(A)
FH = eigen(H)

```

```julia
V = FA.vectors
Λ = Diagonal(FA.values)

A_reconstructed = V * Λ * inv(V)
A ≈ A_reconstructed
```

```julia
V = FH.vectors
Λ = Diagonal(FH.values)
H_reconstructed = V * Λ * V'
H ≈ H_reconstructed
```

```julia
function random_density_matrix(x::Int)
    A = randn(x,x) + im * randn(x,x)
    H = A + A'
    p = H/tr(H)
    while !isposdef(p)
        p += Diagonal(ones(x))
    end
    return p
end 

ρ = random_density_matrix(2)
println(ρ≈ρ')
println(tr(ρ))
println(eigen(ρ).values)
```

```julia
using LinearAlgebra

ρ1 = [1/2 0; 0 1/2]
ρ2 = [1/2 1/2; 1/2 1/2]
ρ3 = [1/2 1; 1 1/2]
ρ4 = [1/3 im/4; -im/4 2/3]

function is_density_matrix(ρ; tol=1e-8)
    hermitian_check = ρ ≈ ρ'
    trace_check = abs(tr(ρ) - 1) < tol
    eigenvalue_check = minimum(real(eigen(ρ).values)) >= -tol
    return hermitian_check && trace_check && eigenvalue_check
end

for ρ in [ρ1, ρ2, ρ3, ρ4]
    println(ρ)
    println("trace = ", tr(ρ))
    println("det = ", det(ρ))
    println("eigenvalues = ", eigen(ρ).values)
    println("density matrix? ", is_density_matrix(ρ))
    println()
end
```
